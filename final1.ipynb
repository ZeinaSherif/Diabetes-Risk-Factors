```python
df['age'] = np.floor(df['age']).astype(int)
```

```python
def categorize_age(age):
    if age <= 2:
        return 'Infants'
    elif age <= 12:
        return 'Children'
    elif age <= 35:
        return 'Teenagers and Young Adults'
    elif age <= 55:
        return 'Middle-aged Adults'
    else:
        return 'Older Adults'
```

```python
# Apply categorization
df['age'] = df['age'].apply(categorize_age)
df['age'].value_counts()
```

```python
age_counts = df['age'].value_counts().sort_index()
# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(age_counts.index, age_counts.values, color='skyblue')
plt.title('Bar Plot of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.xticks(age_counts.index)  # Show all age values on x-axis
plt.show()
```

```python
df['smoking_history'].replace({'not current':'former', 'ever':'former'}, inplace=True)
df['smoking_history'].value_counts()
```

```python
def categorize_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal weight'
    elif 25 <= bmi < 30:
        return 'Overweight'
    else:
        return 'Obesity'
```

```python
df['bmi'] = df['bmi'].apply(categorize_bmi)
df['bmi'].value_counts()
```
